```python
#Question :1
 #class and object with respect to object oriented programming.
 #In object-oriented programming (OOP), a class is a blueprint for creating objects, which are instances of the class. A class defines a set of 
 #attributes (also called properties) and methods that characterize the behavior of objects created from it.

 #For example, imagine a class called "Car" that defines the properties and behaviors of a car, such as its make, model, year, color, and methods
 #like starting the engine, accelerating, and braking.

 #An object, on the other hand, is an instance of a class that has specific values assigned to its properties. For instance, an object of the "Car"
 #class could be an actual car with a make of "Thar" ,a model of "LX4 str" ,a year of "2020", and a color of "Black".
 
 ##For example:-
class Car:
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color
    
    def start_engine(self):
        print("The car engine is starting...")
    
    def accelerate(self):
        print("The car is accelerating...")
    
    def brake(self):
        print("The car is braking...")

# Creating a object for class:-
my_car = Car("Thar", "LX4 str", 2020, "Black")

#so here we create an object of class car which sows information aboyt particular thing.
    
```

```python
#Question :5
 #Inheritance :-
 #Inheritance is a concept in object-oriented programming (OOP) that allows a new class (called the "subclass" or "derived class") to
 #be based on an existing class (called the "base class" or "superclass"), inheriting its properties and methods. This allows 
 #the subclass to reuse code and add or modify its own functionality as needed.

 #types with  example:-
 #1)Single Inheritance: In this type of inheritance, a subclass inherits from a single base class. 
 #example:-
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors
  ## Here "Car" class inherits from the "Vehicle" class 

 #2)Multiple Inheritance: In this type of inheritance, a subclass inherits from two or more base classes.
 #example:-
class Runner:
    def run(self):
        print("I can run!")

class Swimmer:
    def swim(self):
        print("I can swim!")

class Athlete(Runner, Swimmer):
    pass
  ## Here "Athlete" class inherits from both the "Runner" and "Swimmer" classes
    
 #3)Hierarchical Inheritance: In this type of inheritance, multiple subclasses inherit from a single base class.
 #example:-
class Animal:
    def eat(self):
        print("I can eat!")

class Mammal(Animal):
    def walk(self):
        print("I can walk!")

class Bird(Animal):
    def fly(self):
        print("I can fly!")

class Fish(Animal):
    def swim(self):
        print("I can swim!")
  ##  Here "Mammal", "Bird", and "Fish" classes all inherit from the "Animal" class
    
 #4)Multi-level Inheritance: In this type of inheritance, a subclass inherits from a base class, which in turn inherits from 
 #  another base class.
 #example:-
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department

class SeniorManager(Manager):
    def __init__(self, name, emp_id, department, num_employees):
        super().__init__(name, emp_id,department, num_employees)

  ## Here "SeniorManager" inherit from "Manager" and "Manager" is inherit from "Empl

```
